```python
# Step 1: Import NumPy
import numpy as np

# Step 2: Create a 1D array representing years of experience
years_exp = np.array([1, 3, 5, 7, 10])
print("Years of Experience:", years_exp)

# Step 3: Create a 2D array representing sample salaries (in thousands)
salaries = np.array([[50, 60, 70], [80, 90, 100]])
print("Salary Matrix:\n", salaries)

# Step 4: Create an array of zeros and ones for placeholder analysis
zeros_array = np.zeros((2, 2)) # 2x2 zeros
ones_array = np.ones((2, 3)) # 2x3 ones
print("Zeros Array:\n", zeros_array)
print("Ones Array:\n", ones_array)
```

```python
# Step 1: Element-wise addition
exp_plus_5 = years_exp + 5 # Add 5 years to all experience values
print("Years + 5:", exp_plus_5)

# Step 2: Element-wise multiplication
exp_times_2 = years_exp * 2 # Multiply all values by 2
print("Years * 2:", exp_times_2)

# Step 3: Dot product (simulate salary projections)
sample1 = np.array([1, 2, 3])
sample2 = np.array([4, 5, 6])
dot_result = np.dot(sample1, sample2)
print("Dot Product:", dot_result)
```

```python
# Step 1: Access individual element
print("First year of experience:", years_exp[0])

# Step 2: Slice arrays
print("First two salaries:", salaries[0, :2]) # First two elements of first row

# Step 3: Access all rows for a specific column
print("Second column salaries:", salaries[:, 1])

# Step 4: Negative indexing
print("Last year of experience:", years_exp[-1])
```

```python
# Step 1: Reshape 1D array into 2x3 matrix for batch analysis
reshaped_exp = np.reshape(np.arange(1, 7), (2, 3))
print("Reshaped Experience Array:\n", reshaped_exp)

# Step 2: Flatten 2D arrays back into 1D
flattened_exp = reshaped_exp.flatten()
print("Flattened Array:", flattened_exp)

# Step 3: Transpose example
print("Transposed Array:\n", reshaped_exp.T)
```

```python
# Step 1: Add a bonus array to salaries using broadcasting
bonus = np.array([5, 10, 15])
salaries_with_bonus = salaries + bonus # Adds corresponding bonus to each column
print("Salaries after bonus: \n", salaries_with_bonus)
```

```python
# Step 1: Mean of years of experience
print("Mean experience:", np.mean(years_exp))

# Step 2: Standard deviation of experience
print("Std deviation of experience:", np.std(years_exp))

# Step 3: Max and Min salaries
print("Max salary:", np.max(salaries), "Min salary:", np.min(salaries))

# Step 4: Sum of salaries
print("Sum of all salaries:", np.sum(salaries))
```

```python
# Step 1: Apply trigonometric functions
angles = np.array([0, np.pi/4, np.pi/2])
print("Sine of angles:", np.sin(angles))
print("Cosine of angles:", np.cos(angles))

# Step 2: Apply function along rows (sum salaries per person)
salary_sums = np.apply_along_axis(np.sum, 1, salaries)
print("Sum of Salaries per person:", salary_sums)
```

```python
import pandas as pd
import numpy as np

# Step 1: Load the dataset
df_kaggle = pd.read_csv('survey_results_public.csv')

# Step 2: Select relevant columns
df_subset = df_kaggle[['Country', 'EdLevel', 'YearsCodePro', 'ConvertedCompYearly']]

# Step 3: Clean the data
df_clean = df_subset.dropna().copy()

# --- FIX FOR TYPEERROR ---
# This converts text like 'Less than 1 year' to numbers so np.where works
df_clean['YearsCodePro'] = pd.to_numeric(df_clean['YearsCodePro'], errors='coerce')
df_clean = df_clean.dropna() 

# Step 4: Categorize experience
df_clean['ExperienceLevel'] = np.where(df_clean['YearsCodePro'] >= 10, 'Senior', 'Junior')

# Step 5: Group and compute average
grouped_data = df_clean.groupby(['Country', 'ExperienceLevel'])['ConvertedCompYearly'].mean()

# Step 6: Reset index and print
grouped_data = grouped_data.reset_index()
print("Formatted Grouped Data:\n", grouped_data.head())
```

```python
import numpy as np
import pandas as pd

STUDENT_NAME = "PIDO"

STUDENT_ID = 255526
```

```python
# 1: 1D Array Statistics
np.random.seed(STUDENT_ID)
data = np.random.randint(1, STUDENT_ID % 100 + 50, size=10)
print(f"{STUDENT_NAME}'s Array:", data)
mean_val = np.mean(data)
std_val = np.std(data)
print("Mean:", mean_val, "Std Dev:", std_val)
```

```python
# 2: Row and Column Sums
np.random.seed(STUDENT_ID)
matrix = np.random.randint(1, STUDENT_ID % 50 + 20, size=(3,4))
print(f"{STUDENT_NAME}'s Matrix:\n", matrix)
row_sums = np.sum(matrix, axis=1)
col_sums = np.sum(matrix, axis=0)
print("Row sums:", row_sums)
print("Column sums:", col_sums)
```

```python
# 3: 3D Array Slicing
np.random.seed(STUDENT_ID)
arr3d = np.random.randint(0, STUDENT_ID % 20 + 10, size=(3,3,3))
print(f"{STUDENT_NAME}'s 3D Array:\n", arr3d)
slice_mid = arr3d[:,1,:]
print("Slice[:,1,:]:\n", slice_mid)
```

```python
# 4: Boolean Filtering
np.random.seed(STUDENT_ID)
x = np.random.randint(1, STUDENT_ID % 100 + 50, size=10)
filtered = x[(x > STUDENT_ID % 50) & (x < STUDENT_ID % 100)]
print(f"{STUDENT_NAME}'s Original:", x)
print("Filtered subset:", filtered)
```

```python
# 5: Matrix Multiplication
np.random.seed(STUDENT_ID)
A = np.random.randint(1, STUDENT_ID % 50 + 20, size=(2,2))
B = np.random.randint(1, STUDENT_ID % 50 + 20, size=(2,2))
product = A @ B
det_A = np.linalg.det(A)
print(f"{STUDENT_NAME}'s A @ B:\n", product)
print("Determinant of A:", det_A)
```

```python
# 6: Trigonometric Functions
angles = np.linspace(0, 2*np.pi, 8)
sine_vals = np.sin(angles)
cos_vals = np.cos(angles)
print(f"{STUDENT_NAME}'s Angles:", angles)
print("Sine:", sine_vals)
print("Cosine:", cos_vals)
```

```python
# 7: Slicing with Steps
np.random.seed(STUDENT_ID)
arr = np.random.randint(1, STUDENT_ID % 50 + 20, size=(4,4))
arr[::2, ::2] = 0
print(f"{STUDENT_NAME}'s Modified 4x4 Array:\n", arr)
```

```python
# 8: Pass/Fail Logic (np.where)
np.random.seed(STUDENT_ID)
scores = np.random.randint(0, STUDENT_ID % 100 + 50, 10)
grades = np.where(scores >= STUDENT_ID % 70, 'Pass', 'Fail')
print(f"{STUDENT_NAME}'s Scores:", scores)
print("Grades:", grades)
```

```python
# 9: Shape Manipulation
np.random.seed(STUDENT_ID)
arr = np.random.randint(1, STUDENT_ID % 50 + 20, size=(3,4))
print(f"{STUDENT_NAME}'s Flattened:", arr.flatten())
print("Transposed:\n", arr.T)
```

```python
# 10: Even/Odd Masking
np.random.seed(STUDENT_ID)
arr = np.random.randint(1, STUDENT_ID % 50 + 20, (3,3))
mask = arr % 2 == 0
arr[mask] = -1
print(f"{STUDENT_NAME}'s Conditional Array:\n", arr)
```

```python
# 11: DataFrame Creation
np.random.seed(STUDENT_ID)
df = pd.DataFrame({
    'Name': [STUDENT_NAME]*5,
    'Score': np.random.randint(STUDENT_ID%50, STUDENT_ID%100+50, 5),
    'YearsCodePro': np.random.randint(0, STUDENT_ID%20+10, 5)
})
print(df)
```

```python
# 12: Grouped Salary Averages
np.random.seed(STUDENT_ID)
df = pd.DataFrame({
    'EdLevel': ['Bachelor','Master','PhD','Bachelor','Master'],
    'ConvertedComp': np.random.randint(40000, 150000, 5)
})
print(df.groupby('EdLevel')['ConvertedComp'].mean())
```

```python
# 13: Sorting Values
np.random.seed(STUDENT_ID)
df = pd.DataFrame({
    'Country': ['Philippines','USA','UK','Canada','Germany'],
    'ConvertedComp': np.random.randint(40000, 150000, 5)
})
print(df.sort_values(by='ConvertedComp', ascending=False))
```

```python
# 14: Experience Filter
np.random.seed(STUDENT_ID)
df = pd.DataFrame({
    'Name':[STUDENT_NAME]*5,
    'YearsCodePro': np.random.randint(0, STUDENT_ID%20+10, 5)
})
print(df[df['YearsCodePro'] > STUDENT_ID % 10])
```

```python
# 15: Bracket Binning
np.random.seed(STUDENT_ID)
df = pd.DataFrame({'ConvertedComp': np.random.randint(40000, 150000, 5)})
bins = [0, 50000, 100000, 150000, 200000]
labels = ['0-50k', '50-100k', '100-150k', '>150k']
df['Bracket'] = pd.cut(df['ConvertedComp'], bins=bins, labels=labels)
print(df)
```

```python
# 16: Logarithmic Scaling
np.random.seed(STUDENT_ID)
df = pd.DataFrame({'ConvertedComp': np.random.randint(40000, 150000, 5)})
df['LogComp'] = np.log(df['ConvertedComp'])
print(df)
```

```python
# 17: Correlation Matrices
np.random.seed(STUDENT_ID)
df = pd.DataFrame({
    'YearsCodePro': np.random.randint(0, STUDENT_ID%20+10, 5),
    'ConvertedComp': np.random.randint(40000, 150000, 5)
})
print(df.corr())
```

```python
# 18: Multiple Aggregations
np.random.seed(STUDENT_ID)
df = pd.DataFrame({
    'EdLevel':['Bachelor','Master','PhD','Bachelor','Master'],
    'ConvertedComp': np.random.randint(40000, 150000, 5)
})
summary = df.groupby('EdLevel')['ConvertedComp'].agg(['mean','median','std']).reset_index()
print(summary)
```

```python
# 19: High Pay Classification
np.random.seed(STUDENT_ID)
df = pd.DataFrame({'ConvertedComp': np.random.randint(40000, 150000, 5)})
df['HighPay'] = np.where(df['ConvertedComp'] > 110000, 'Yes', 'No')
print(df)
```

```python
# 20: Country Max/Mean Summary
np.random.seed(STUDENT_ID)
df = pd.DataFrame({
    'Country': ['Philippines','USA','UK','Canada','Germany'],
    'ConvertedComp': np.random.randint(40000, 150000, 5)
})
print(df.groupby('Country')['ConvertedComp'].agg(['mean','max']))
```

```python
import numpy as np
import pandas as pd

# Detecting NaNs and Invalid Data in Arrays (NumPy)
# This removes np.nan (Not a Number) and np.inf (Infinite) values
data = np.array([1, 2, np.nan, 4, np.inf])
cleaned = data[np.isfinite(data)] 
print("Cleaned NumPy Data:", cleaned)

# Handling Missing Data in Pandas
# dropna() removes rows containing None or NaN values
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", None], 
    "Score": [85, None, 92, 88]
})
df_cleaned = df.dropna()
print("\nCleaned DataFrame:\n", df_cleaned)
```
